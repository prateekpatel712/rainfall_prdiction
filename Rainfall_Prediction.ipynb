```python
#Plotting the graph
day_index = 257
day = [i for i in range(y.size)]
```

```python
plt.title("PREDICTION TREND GRAPH")
plt.scatter(day, y, c='g')
plt.scatter(day[day_index], y[day_index], c='r')
plt.xlabel("Days")
plt.ylabel("Prediction in inches")    
plt.legend()
plt.show;
```

```python
x_vis = x.filter(["TempAvgF", "DewPointAvgF", "HumidityAvgPercent", "SeaLevelPressureAvgInches", "VisibilityAvgMiles", "WindAvgMPH"]);
```

```python
print("Prediction vs attribute trend graph")
for i in range(x_vis.columns.size):
    plt.subplot(4,2, i+1)
    plt.scatter(day, x_vis[x_vis.columns.values[i][:100]], c='g')
    plt.scatter(day[day_index], x_vis[x_vis.columns.values[i]][day_index], c='r')
    plt.title(x_vis.columns.values[i])
plt.show();       
```
